# Matplotlib

Matplotlib is a library for making 2D plots of arrays in Python.  Although Matplotlib is written primarily in pure Python, it makes heavy use of NumPy and other extension code to provide good performance even for large arrays. Matplot lib contain a lot of functionality but we only focus on some of the parts of matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(start=0,stop=10,step=1)
y = x**2
```

## Line
Line graph is used for analysing continuous data. 

```python
# line plotting
plt.plot(x,y,'r--')
plt.ylabel('Y Axis Name')
plt.xlabel('X Axis Name')
plt.show()
```

```python
# Zoom
plt.plot(x,y,'r--')
plt.xlim(0,4)
plt.ylim(0,4)
plt.show()
```

```python
# ploting sine wave
x = np.arange(0, 5, 0.1);
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

## Point

```python
# data point plotting
data_x = [1,2,3,4,5,6]
data_y = [1,1,2,3,5,8]
plt.plot(data_x, data_y, 'ro')
plt.axis([0, 7, 0, 10])
plt.show()
```

## Bar Graph 

```python
# simple bar chart 
x_label = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
x_position = np.arange(len(x_label))
data_y = [10,8,6,4,2,1]
plt.bar(x_position, data_y, align='center', alpha=0.5)
plt.xticks(x_position, x_label)
plt.title('Programming language')
plt.show()
```

## Pie Chart

```python
# pie chart
label = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
ratio = [10,8,6,4,2,1]
plt.pie(ratio, labels = label, startangle=90)
plt.axis('equal')
plt.show()
```

```python
matrix = np.arange(0,100).reshape(10,10)
plt.imshow(matrix);
```

```python
matrix = np.random.randint(0,100,(10,10))
plt.imshow(matrix);
```

```python
plt.imshow(matrix,cmap='coolwarm')
plt.colorbar();
```
